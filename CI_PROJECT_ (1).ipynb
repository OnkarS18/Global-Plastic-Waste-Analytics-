```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
X = df.drop("Plastic_Waste_Level", axis=1)
y = df["Plastic_Waste_Level"]

# -----------------------------
#  Encode Categorical Columns
# -----------------------------
label_encoder = LabelEncoder()

# Encode target column (output)
y = label_encoder.fit_transform(y)

# Encode categorical input columns
X["Country"] = label_encoder.fit_transform(X["Country"])
X["Main_Sources"] = label_encoder.fit_transform(X["Main_Sources"])

# -----------------------------
#  Standardize Numerical Features
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -----------------------------
#  Train-Test Split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# -----------------------------
#  Print Dataset Shapes
# -----------------------------
print(f"Shape of training data : {X_train.shape}, {y_train.shape}")
print(f"Shape of testing data  : {X_test.shape}, {y_test.shape}")
```

```python
#  Import Libraries
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
#  Train the Logistic Regression model
# -----------------------------
lr = LogisticRegression(max_iter=1000)   # increase max_iter for convergence
lr.fit(X_train, y_train)

# -----------------------------
#  Make predictions
# -----------------------------
lr_pred = lr.predict(X_test)

# -----------------------------
#  Generate confusion matrix
# -----------------------------
lr_conf = confusion_matrix(y_test, lr_pred)

# -----------------------------
#  Generate classification report
# -----------------------------
lr_report = classification_report(y_test, lr_pred)

# -----------------------------
#  Calculate accuracy
# -----------------------------
lr_acc = round(accuracy_score(y_test, lr_pred) * 100, 2)

# -----------------------------
#  Print results
# -----------------------------
print(f"Confusion Matrix: \n\n{lr_conf}")
print(f"\nClassification Report: \n\n{lr_report}")
print(f"\nThe Accuracy of Logistic Regression is {lr_acc} %")

# -----------------------------
#  Plot confusion matrix as heatmap
# -----------------------------
plt.figure(figsize=(8, 7))
sns.set(font_scale=1.4)

# Get class labels (in case you want actual category names)
import numpy as np
classes = np.unique(y_test)

sns.heatmap(
    lr_conf,
    annot=True,
    fmt='g',
    cmap='Blues',
    cbar=True,
    annot_kws={"size": 16, "weight": 'bold'},
    xticklabels=classes,
    yticklabels=classes,
    linewidths=1.5,
    linecolor='white',
    cbar_kws={"shrink": 0.8}
)

plt.title('Confusion Matrix - Logistic Regression', fontsize=18, fontweight='bold')
plt.xlabel('Predicted Labels', fontsize=14)
plt.ylabel('Actual Labels', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#  Import Libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# ----------------------------------------
#  Train the K-Nearest Neighbors Model
# ----------------------------------------
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# ----------------------------------------
#  Make Predictions
# ----------------------------------------
knn_pred = knn.predict(X_test)

# ----------------------------------------
#  Evaluation Metrics
# ----------------------------------------
knn_conf = confusion_matrix(y_test, knn_pred)
knn_report = classification_report(y_test, knn_pred)
knn_acc = round(accuracy_score(y_test, knn_pred) * 100, 2)

# ----------------------------------------
#  Print Results
# ----------------------------------------
print(f"Confusion Matrix: \n\n{knn_conf}")
print(f"\nClassification Report: \n\n{knn_report}")
print(f"\nThe Accuracy of K-Nearest Neighbors Classifier is {knn_acc} %")

# ----------------------------------------
#  Confusion Matrix Heatmap
# ----------------------------------------
plt.figure(figsize=(8, 7))
sns.set(font_scale=1.4)

sns.heatmap(knn_conf, annot=True, fmt='g', cmap='Greens', cbar=True,
            annot_kws={"size": 16, "weight": 'bold'},
            xticklabels=['Predicted 0', 'Predicted 1', 'Predicted 2'],
            yticklabels=['Actual 0', 'Actual 1', 'Actual 2'],
            linewidths=1.5, linecolor='white', cbar_kws={"shrink": 0.8})

plt.title('Confusion Matrix - K Nearest Neighbors (Plastic Waste Level)', fontsize=18, fontweight='bold')
plt.xlabel('Predicted Labels', fontsize=14)
plt.ylabel('Actual Labels', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#  Import Libraries
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# ----------------------------------------
#  Train the XGBoost Classifier (Tuned)
# ----------------------------------------
xgb = XGBClassifier(
    n_estimators=300,          # number of boosting rounds
    learning_rate=0.05,        # smaller  more accurate but slower
    max_depth=8,               # controls model complexity
    subsample=0.8,             # use 80% data per tree (reduces overfitting)
    colsample_bytree=0.8,      # use 80% features per tree
    gamma=0.2,                 # regularization term
    reg_lambda=1,              # L2 regularization
    random_state=42,
    objective='multi:softmax', # multiclass classification
    num_class=3                # since 0,1,2  Low, Medium, High
)

# ----------------------------------------
#  Train the Model
# ----------------------------------------
xgb.fit(X_train, y_train)

# ----------------------------------------
#  Make Predictions
# ----------------------------------------
xgb_pred = xgb.predict(X_test)

# ----------------------------------------
#  Evaluate Performance
# ----------------------------------------
xgb_conf = confusion_matrix(y_test, xgb_pred)
xgb_report = classification_report(y_test, xgb_pred)
xgb_acc = round(accuracy_score(y_test, xgb_pred) * 100, 2)

# ----------------------------------------
#  Print Results
# ----------------------------------------
print(f"Confusion Matrix: \n\n{xgb_conf}")
print(f"\nClassification Report: \n\n{xgb_report}")
print(f"\n The Accuracy of XGBoost Classifier is {xgb_acc} %")

# ----------------------------------------
#  Confusion Matrix Heatmap
# ----------------------------------------
plt.figure(figsize=(8, 7))
sns.set(font_scale=1.3)

sns.heatmap(xgb_conf, annot=True, fmt='g', cmap='Purples', cbar=True,
            annot_kws={"size": 14, "weight": 'bold'},
            xticklabels=['Predicted 0', 'Predicted 1', 'Predicted 2'],
            yticklabels=['Actual 0', 'Actual 1', 'Actual 2'],
            linewidths=1.5, linecolor='white', cbar_kws={"shrink": 0.8})

plt.title('Confusion Matrix - XGBoost Classifier (Plastic Waste Level)', fontsize=18, fontweight='bold')
plt.xlabel('Predicted Labels', fontsize=14)
plt.ylabel('Actual Labels', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#  Import Libraries
from sklearn.ensemble import VotingClassifier, RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# ----------------------------------------
#  Define Base Models
# ----------------------------------------
knn = KNeighborsClassifier(n_neighbors=3)
svm = SVC(C=100, gamma=0.002, kernel='rbf', probability=True)
rf = RandomForestClassifier(n_estimators=300, max_depth=15, criterion='entropy', random_state=42)
xgb = XGBClassifier(
    n_estimators=300, learning_rate=0.05, max_depth=8,
    subsample=0.8, colsample_bytree=0.8, gamma=0.2,
    reg_lambda=1, random_state=42, objective='multi:softmax', num_class=3
)
log = LogisticRegression(max_iter=2000, solver='lbfgs', multi_class='auto', random_state=42)

# ----------------------------------------
#  Create Voting Classifier
# ----------------------------------------
voting_clf = VotingClassifier(
    estimators=[
        ('KNN', knn),
        ('SVM', svm),
        ('RF', rf),
        ('XGB', xgb),
        ('LOG', log)
    ],
    voting='soft'   # 'soft'  uses predicted probabilities (more accurate)
)

# ----------------------------------------
#  Train the Combined Model
# ----------------------------------------
voting_clf.fit(X_train, y_train)

# ----------------------------------------
#  Make Predictions
# ----------------------------------------
voting_pred = voting_clf.predict(X_test)

# ----------------------------------------
#  Evaluate Performance
# ----------------------------------------
voting_conf = confusion_matrix(y_test, voting_pred)
voting_report = classification_report(y_test, voting_pred)
voting_acc = round(accuracy_score(y_test, voting_pred) * 100, 2)

# ----------------------------------------
#  Print Results
# ----------------------------------------
print(f"Confusion Matrix: \n\n{voting_conf}")
print(f"\nClassification Report: \n\n{voting_report}")
print(f"\n The Accuracy of 5-Model Voting Classifier is {voting_acc} %")

# ----------------------------------------
#  Confusion Matrix Heatmap
# ----------------------------------------
plt.figure(figsize=(8, 7))
sns.set(font_scale=1.3)

sns.heatmap(voting_conf, annot=True, fmt='g', cmap='Greens', cbar=True,
            annot_kws={"size": 14, "weight": 'bold'},
            xticklabels=['Predicted 0', 'Predicted 1', 'Predicted 2'],
            yticklabels=['Actual 0', 'Actual 1', 'Actual 2'],
            linewidths=1.5, linecolor='white', cbar_kws={"shrink": 0.8})

plt.title('Confusion Matrix - Voting Classifier (Plastic Waste Level)', fontsize=18, fontweight='bold')
plt.xlabel('Predicted Labels', fontsize=14)
plt.ylabel('Actual Labels', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# ------------------------------------------------------
# FINAL MODEL TRAINING + SAVE BEST MODEL
# ------------------------------------------------------

import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import pandas as pd

# -----------------------------
# PREPROCESSING FUNCTION
# -----------------------------
def preprocess_for_model(df):
    df = df.copy()
    target = "Plastic_Waste_Level"

    drop_cols = ["Country", target]
    feature_cols = [c for c in df.columns if c not in drop_cols]

    num_cols = df[feature_cols].select_dtypes(include=['number']).columns.tolist()
    cat_cols = [c for c in feature_cols if c not in num_cols]

    # Numeric part
    X_num = df[num_cols].astype(float).reset_index(drop=True)

    # Categorical part
    X_cat = df[cat_cols].fillna("Unknown").reset_index(drop=True)

    encoder = None
    if len(cat_cols) > 0:
        encoder = OrdinalEncoder()
        X_cat_enc = encoder.fit_transform(X_cat)
        X_cat_enc = pd.DataFrame(X_cat_enc, columns=cat_cols)
        X = pd.concat([X_num, X_cat_enc], axis=1)
    else:
        X = X_num

    y = df[target]

    return X, y, encoder, num_cols, cat_cols


# -----------------------------
# DATA SPLIT
# -----------------------------
X, y, encoder, num_cols, cat_cols = preprocess_for_model(df)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)


# -----------------------------
# DEFINE MODELS TO TEST
# (add more models anytime)
# -----------------------------
models = {
    "RandomForest": RandomForestClassifier(
        n_estimators=150,
        max_depth=10,
        random_state=42
    )
    # You can easily add:
    # "LogisticRegression": LogisticRegression(max_iter=500),
    # "XGBoost": XGBClassifier()
}


# -----------------------------
# TRAIN + EVALUATE ALL MODELS
# -----------------------------
results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    results[name] = (acc, model)

    print(f"\nModel: {name}")
    print(f"Accuracy: {acc:.4f}")
    print(classification_report(y_test, preds))


# -----------------------------
# SELECT BEST MODEL
# -----------------------------
best_model_name = max(results, key=lambda k: results[k][0])
best_acc, best_model = results[best_model_name]

print("\n---------------------------------------")
print(f"BEST MODEL: {best_model_name}  (Accuracy: {best_acc:.4f})")
print("---------------------------------------")


# -----------------------------
# SAVE BEST MODEL PACKAGE
# -----------------------------
model_package = {
    "model": best_model,
    "encoder": encoder,
    "num_cols": num_cols,
    "cat_cols": cat_cols
}

SAVE_PATH = "best_plastic_waste_model.joblib"
joblib.dump(model_package, SAVE_PATH)

print(f"\nBest model saved successfully to:  {SAVE_PATH}")
```
